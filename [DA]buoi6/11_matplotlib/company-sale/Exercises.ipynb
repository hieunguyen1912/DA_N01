# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
# read in the data
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
df.head()
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
plt.plot(df['month_number'], df['total_profit'])
plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python

plt.plot(
    df['month_number'],
    df['total_profit'],
    color='r',
    linewidth=3,
    marker='o',
    markerfacecolor='black',
    markersize=8,
    linestyle='--'
)
plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
plt.plot(df['month_number'], df['facecream'], label='Face Cream', marker='o', linewidth=3)
plt.plot(df['month_number'], df['facewash'], label='Face Wash', marker='o', linewidth=3)
plt.plot(df['month_number'], df['toothpaste'], label='Toothpaste', marker='o', linewidth=3)
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap', marker='o', linewidth=3)
plt.plot(df['month_number'], df['shampoo'], label='Shampoo', marker='o', linewidth=3)
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer', marker='o', linewidth=3)

plt.title('Sales data')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
import matplotlib.ticker as ticker

plt.scatter(df['month_number'], df['toothpaste'], color='blue', label='Toothpaste', marker='o')
plt.title('Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.legend()
plt.grid(True, linestyle='--')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
import numpy as np

bar_width = 0.4
months = np.array(df['month_number'])

plt.bar(months - bar_width/2, df['facecream'], width=bar_width, color='skyblue', label='Face Cream', alpha=0.7)
plt.bar(months + bar_width/2, df['facewash'], width=bar_width, color='orange', label='Face Wash', alpha=0.7)

plt.title('Face Cream and Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.legend()
plt.grid(True, linestyle='--')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
plt.bar(df['month_number'], df['bathingsoap'], color='blue')
plt.title('Bathing Soap Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.grid(True, linestyle='--')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
plt.hist(df['total_profit'], bins=7, color='dodgerblue', label='Profit data')
plt.title('Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
sales = [
    df['facecream'].sum(),
    df['facewash'].sum(),
    df['toothpaste'].sum(),
    df['bathingsoap'].sum(),
    df['shampoo'].sum(),
    df['moisturizer'].sum()
]
colors = ['#3498db', '#ff9900', '#2ecc40', '#e74c3c', '#9b59b6', '#8d6748']

plt.pie(sales, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Sales data')
plt.legend(labels, loc='best')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
plt.subplot(2, 1, 1) 
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap', marker='o', linewidth=3, color='black')
plt.title('Bathing Soap Sales Data')

plt.subplot(2, 1, 2)
plt.plot(df['month_number'], df['facewash'], label='facewash', marker='o', linewidth=3, color='blue')
plt.title('Facewash Sales Data')

plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.stackplot(df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer'], labels=labels, colors = colors)
plt.legend()
plt.title('All product sales data using stackplot')
plt.xlabel('Month Number')
plt.ylabel('Salse units in number')
plt.show()
```
